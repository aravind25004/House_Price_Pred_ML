```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
id = test_dataset['Id']
test_ids = id.values
```

```python
selected_columns = ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'SalePrice']
train_dataset = train_dataset[selected_columns]
print(train_dataset.head())
```

```python
x_train = train_dataset.iloc[:, :-1].values
y_train = train_dataset.iloc[:, -1].values
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred_train = regressor.predict(x_train)
print(y_pred_train)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=regressor, X=x_train, y=y_train, cv=10)
print(accuracies.mean())
print(accuracies.std())
```

```python
selected_columns_test = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
test_datasets = test_dataset[selected_columns_test]
print(test_datasets.head())
```

```python
x_test = test_datasets.iloc[:, :].values
print(x_test)
```

```python
y_pred_test = regressor.predict(x_test)
print(y_pred_test)
```

```python
predictions = pd.DataFrame({'Id': test_ids, 'SalePrice': y_pred_test})
print(predictions)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_train, y_pred_train))
```

```python
predictions.to_csv('Prediction.csv', index=False)
```
